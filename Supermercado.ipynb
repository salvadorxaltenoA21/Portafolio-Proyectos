##Lista de supermercado
Este programa es un menú interactivo para administrar un negocio, usando un diccionario de datos, funciones, listas.

El codido realizá:

*   Mostrar inventario
*   Agregar productos al carrito
*   Eliminar productos del carrito
*   Ver carrito actual
*   Mostrar total a pagar
*   Ajustes de Usuario donde se accede con usuario y contraseña(Agregar inventario nuevo, Eliminar inventario)
*   Salir del menú




##diccionario de datos

```python
inventario = {"manzana":{"cantidad": 50,"precio": 30}, "naranja":{"cantidad": 50,"precio": 12}, "platano":{"cantidad": 50,"precio": 15}, "leche":{"cantidad": 50,"precio": 18}, "queso":{"cantidad": 50,"precio": 24}}

inventario.items()
```

##Funciones


```python


def mostrar_inventario(inventario):
  print("== Inventario ==")
  for producto, precio in inventario.items():
    print(f"{producto}: {precio}")


mostrar_inventario(inventario)
```

```python
carrito = {}

def agregar_al_carrito(inventario, carrito):
  print("== Agregar al carrito ==")
  producto = input("Ingresa lo que quieres agregar: ")

  if producto.lower() in inventario:
    try:
     cantidad = int(input("Ingresa la cantidad: "))
     if cantidad > inventario[producto]["cantidad"]:
        print(f"No hay suficiente {producto}. Solo hay {inventario[producto]['cantidad']} disponible(s).")
     else:
        if producto in carrito:
            carrito[producto] += cantidad
        else:
            carrito[producto] = cantidad

            inventario[producto]["cantidad"] -= cantidad
            print(f"{cantidad} {producto}(s) agregado(s) al carrito.")

    except ValueError:
            print("Por favor, ingresa un número válido para la cantidad.")
  else:
    print("Producto no encontrado")



def eliminar_del_carrito(carrito):
  print("== Eliminar del carrito ==")
  producto = input("Ingresa lo que quieres eliminar")
  if producto in carrito:
    cantidad = carrito[producto]
    del carrito[producto]
    inventario[producto]["cantidad"] += cantidad
    print(f"{cantidad} {producto}(s) regresado(s) al inventario y eliminado(s) del carrito.")
  else:
    print("Producto no encontrado")

def ver_carrito(carrito):
  print("== Mostrar carrito ==")
  for producto, cantidad in carrito.items():
    print(f"El producto es {producto} con la cantidad {cantidad}")


def precio_total_producto(inventario,carrito):
    total=0
    print("Detalle del carrito:\n")

    for articulo, cantidad in carrito.items():
        precio = inventario[articulo]["precio"]
        subtotal = cantidad * precio
        total += subtotal
        print(f"{articulo.capitalize()}: {cantidad} x ${precio:.2f} = ${subtotal:.2f}")

    print(f"\nTotal del carrito: ${total:.2f}")
    return total


def ajuste_usuario(inventario):
  print("Ajustes del usuario: \n")
  usuario="admin07"
  contra="210702"
  intentos=3
  while intentos > 0:
    print(f"\nIntentos restantes: {intentos}")
    U_ins = input("Ingrese su usuario: ")
    C_ins = input("Ingresa la contraseña: ")

    if U_ins == usuario and C_ins == contra:
        print("\n¡Acceso concedido! Bienvenido.")
        mostrar_inventario(inventario)

        # Menú principal
        while True:
            print("\n¿Qué desea hacer?")
            print("1. Agregar productos al inventario")
            print("2. Quitar productos del inventario")
            print("3. Mostrar productos del inventario")
            print("4. Salir")

            op = input("Seleccione una opción: ")

            try:
                op = int(op)  # Convertir la opción a entero
            except ValueError:
                print("Por favor, ingrese un número válido.")
                continue

            if op == 1:
                # Opción 1: Agregar productos al inventario
                print("\n== Agregar Artículo al Inventario ==")
                producto = input("Ingresa el nombre del producto: ").lower()

                if producto in inventario:
                    print(f"El producto '{producto}' ya existe en el inventario.")
                else:
                    try:
                        cantidad = int(input("Ingresa la cantidad disponible: "))
                        precio = float(input("Ingresa el precio del producto: "))

                        inventario[producto] = {
                            "cantidad": cantidad,
                            "precio": precio,

                        }

                        print(f"Producto '{producto}' agregado exitosamente al inventario.")
                    except ValueError:
                        print("Error: Ingresa valores numéricos válidos para la cantidad y el precio.")

            elif op == 2:
                # Opción 2: Eliminar productos del inventario
                print("\n== Eliminar Artículo del Inventario ==")
                producto = input("Ingresa el nombre del producto que deseas eliminar: ").lower()

                if producto in inventario:
                    del inventario[producto]  # Eliminar el producto del diccionario
                    print(f"El producto '{producto}' ha sido eliminado del inventario.")
                else:
                    print(f"El producto '{producto}' no se encuentra en el inventario.")

            elif op == 3:
              mostrar_inventario(inventario)
            elif op == 4:
                # Opción 4: Salir
                print("\nSaliendo...")
                break

            else:
                print("Por favor, selecciona una opción válida (1, 2, , 3 o 4).")
        break  # Salir del programa si se accede correctamente
    else:
        print("Usuario o contraseña incorrectos.")
        intentos -= 1

  if intentos == 0:
    print("\nHas agotado todos los intentos. Acceso denegado.")



```

##Menú principal

```python


while True:
  print("== Menu principal ==")
  print("1. Mostrar inventario")
  print("2. Agregar productos al carrito")
  print("3. Eliminar productos del carrito")
  print("4. Ver carrito actual")
  print("5. Mostrar total a pagar")
  print("6. Ajustes de Usuario")
  print("7. Salir")


  opcion = input("Seleccione una opción: ")
  opcion = int(opcion)

  if opcion == 1:
    mostrar_inventario(inventario)

  elif opcion == 2:
    agregar_al_carrito(inventario, carrito)

  elif opcion == 3:
    eliminar_del_carrito(carrito)

  elif opcion == 4:
    ver_carrito(carrito)

  elif opcion == 5:
    precio_total_producto(inventario=inventario, carrito=carrito)
  elif opcion == 6:
    ajuste_usuario(inventario)
  elif opcion == 7:
    print("Gracias por su visita")
    break
```
